```python
import numpy as np
import matplotlib.pyplot as plt
from GPy.models import BayesianGPLVM
from sklearn.datasets import make_moons, make_swiss_roll, make_circles, make_s_curve


def fit_nd_plot_gplvm(Y):
    model = BayesianGPLVM(Y, 2, init='PCA')
    model.optimize_restarts(15, optimizer='lbfgs')


    E_ygY, V_ygY = model.predict(np.random.normal(size=(500, 2)))
    Y_samp = E_ygY + np.random.normal(0, np.sqrt(V_ygY))
    plt.scatter(Y_samp[:, 0], Y_samp[:, 1], label='sample')
    plt.scatter(Y[:, 0], Y[:, 1], alpha=0.3, label='data')
    plt.legend()

np.random.seed(42); plt.ion(); plt.style.use('ggplot')
n = 500
```

```python
Y_moons, c = make_moons(n, noise=0.1)
fit_nd_plot_gplvm(Y_moons)
```

```python
from GPy.kern import OU

kern = OU(2)
```

```python
Y_spiral, c =  make_swiss_roll(n_samples=n, noise=1.2)
Y_spiral = Y_spiral[:,[0,2]]

fit_nd_plot_gplvm(Y_spiral)
```

```python
Y_circles, c = make_circles(n, factor=0.3, noise=0.03)
fit_nd_plot_gplvm(Y_circles)
```

```python
def make_checkeboard(npar):
    x1 = np.random.rand(npar) * 4 - 2
    x2_ = np.random.rand(npar) - np.random.randint(0, 2, npar) * 2
    x2 = x2_ + (np.floor(x1) % 2)
    x = np.concatenate([x1[:, None], x2[:, None]], 1) * 7.5
    return x
```

```python
Y_board = make_checkeboard(n)
fit_nd_plot_gplvm(Y_board)
```

```python
Y_scurve, c =  make_s_curve(n_samples=n, noise=0.1)
Y_scurve = Y_scurve[:,[0,2]]

fit_nd_plot_gplvm(Y_scurve)
```
